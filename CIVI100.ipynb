**Source:** https://link.springer.com/chapter/10.1007/978-3-319-46466-4_13

## Overview
The **SD-198** dataset is a large-scale collection of **clinical skin disease images** designed for benchmarking computer vision models in dermatological diagnosis. It contains **6,584 images** across **198 different disease categories**, making it one of the most comprehensive publicly available datasets for skin disease classification. A subset, **SD-128**, includes only classes with more than 20 samples to mitigate class imbalance.

## Structure
- **Full dataset (SD-198):** 6,584 images  
- **Subset (SD-128):** Classes with >20 images  
- **Number of classes:** 198 total  
- **Data source:** Images were collected from **DermQuest**, an online dermatology resource where images are submitted by patients and dermatologists.  
- **Annotations:** Each image is labeled and verified by dermatology experts, with additional review by two professionals to ensure quality.

The main folder dataset contains a subfolder and two text files: images, where the classes and images are stored, and the train and test text files which contains subdirectories of the images that are used in trianing and testing.

```python
# Basic data handling
import os
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Image handling
from PIL import Image
from tqdm import tqdm
```

```python
data_dir = '/content/SD-128/images'

image_paths = []
classes = []

# Get class names from the directories
class_names = [d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))]

# Collect all image paths and assign classes by walking through the directories
for class_name in class_names:
    class_dir = os.path.join(data_dir, class_name)
    if os.path.isdir(class_dir):
        for image_name in os.listdir(class_dir):
            if image_name.endswith(('.jpg', '.jpeg', '.png')): # Add other image extensions if necessary
                image_full_path = os.path.join(class_dir, image_name)
                image_paths.append(image_full_path)
                classes.append(class_name)

# Create a DataFrame
df = pd.DataFrame({'image_path': image_paths, 'class': classes})

display(df.head())
```

```python
print("Total images:", len(df))
print("Total classes:", df['class'].nunique())
print("\nAverage images per class:", len(df)/df['class'].nunique())

df['class'].value_counts().describe()
```

Thanks to that short surface-level description of the dataset, we cna confirm and concur a few things:

- **Total images:** This confirms that the DataFrame contains 5619 images.
- **Total classes:** This shows that there are 128 unique disease categories in the SD-128 subset.
- **Average images per class:** This indicates that, on average, each class has about 43.9 images.
- **Descriptive statistics of class counts:** The `describe()` output for `df['class'].value_counts()` gives a more detailed picture of the class distribution:
    - **count:** There are 128 classes represented in the statistics.
    - **mean:** The average number of images per class is 43.89.
    - **std:** The standard deviation of 15.42 shows the spread of image counts across classes.
    - **min:** The minimum number of images in any class is 20, which aligns with the description of the SD-128 subset only including classes with more than 20 samples.
    - **25%, 50% (median), 75%:** These percentiles show the distribution of class sizes. For example, 50% of the classes have 46 or fewer images.
    - **max:** The maximum number of images in any single class is 60.

These statistics help us understand the size and class balance of the SD-128 dataset, which is an important step in the EDA process.
